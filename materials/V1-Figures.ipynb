# Figure for statistical analyses lecture material

```python
from math import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
ntests=np.arange(1,1000) #number of comparisons performed
alphas=[0.05,0.025,0.01,0.001,0.0001]
markers=['k-','k--','k:','b-','b:']
def fwer(alpha,ntests): #probability of a type I error
    fw=[]
    for n in range(len(ntests)):
        fw.append(1-(1-alpha)**ntests[n])
    return fw

plt.figure(figsize=(15,8))
for a in range(len(alphas)):
    fw=fwer(alphas[a],ntests)
    plt.plot(ntests, fw, markers[a],label="alpha={}".format(alphas[a]))
    
plt.legend()
plt.xlabel("Number of comparisons")
plt.ylabel("Type-I error probability")
```
